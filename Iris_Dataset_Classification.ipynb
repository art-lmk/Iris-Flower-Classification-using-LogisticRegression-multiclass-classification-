```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as LogisticRegression
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#load the dataset
df = pd.read_csv('IRIS.csv')
df.head(10)
df.shape
df.info()
```

```python
#scatterplot
sns.FacetGrid(df,hue='species',height=5,aspect=1).map(plt.scatter,'sepal_length','sepal_width').add_legend()
```

```python
#mapping
flower_mapping={'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
df['species']=df['species'].map(flower_mapping)
df.head()
```

```python
#prepare inputs and outputs
x=df[['sepal_length','sepal_width','petal_length','petal_width']].values
y= df[['species']].values
print (y)
```

```python
#model
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x,y)
model.score(x, y)
```

```python
expected=y
predicted=model.predict(x)
predicted
```

```python
#model evaluation
from sklearn import metrics
print(metrics.classification_report(expected,predicted))
print(metrics.confusion_matrix(expected,predicted))
```

In the above output, precision score for 0(Setosa) is 1.00,meaning its a great prediction
for 1(Versicolor) is 0.98 and for 2(Virginica) is 0.94. The model has classified someVersicolor species as Virginica and Vice versa as seanin the scatter plot displayed in Cell 3
